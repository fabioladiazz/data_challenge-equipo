```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Define la paleta de colores
color_palette = ['#1D2A62', '#87AECE', '#B0B7C3', '#E8E4D9', '#BAC7B0', '#AFD06E', '#437118']
background_color = "#ffffff"
```

```python
df = pd.read_csv('df_final.csv')
```

```python
# Convertir las fechas a formato datetime
df['fecha_adj_firme'] = pd.to_datetime(df['fecha_adj_firme'])

# Crear una columna para el período (antes, durante y después de la pandemia)
df['periodo'] = pd.cut(df['fecha_adj_firme'].dt.year,
                       bins=[2017, 2019, 2021, df['fecha_adj_firme'].dt.year.max()],
                       labels=[0, 1, 2])

# Filtrar el DataFrame para que solo queden los datos con las etiquetas especificadas
df = df[df['periodo'].notna()]
```

```python
df['fecha_cierre_recepcion'] = pd.to_datetime(df['fecha_cierre_recepcion'])
df['fecha_adj_firme'] = pd.to_datetime(df['fecha_adj_firme'])

# Calcular la diferencia en días y crear la nueva columna 'duracion'
df['duracion_proc'] = (df['fecha_adj_firme'] - df['fecha_cierre_recepcion']).dt.days
```

```python
df_periodo_0 = df[df['periodo'] == 0]
df_periodo_1 = df[df['periodo'] == 1]
df_periodo_2 = df[df['periodo'] == 2]
```

## General

```python
df['fecha_adj_firme'] = pd.to_datetime(df['fecha_adj_firme'])

# Filtrar los datos para los años 2018 a 2024
df_filtered = df[(df['fecha_adj_firme'].dt.year >= 2018) & (df['fecha_adj_firme'].dt.year <= 2024)]

# Contar la cantidad de datos por año
data_per_year = df_filtered['fecha_adj_firme'].dt.year.value_counts().sort_index()

# Imprimir los resultados
print("Cantidad de datos por año (2018-2023):")
print(data_per_year)

print("Cantidad de datos por año (2018-2023):")
```

```python
df_filtered.loc[:, 'year'] = df_filtered['fecha_adj_firme'].dt.year
```

```python
procedures_per_year = df_filtered['year'].value_counts().sort_index()

# Creación del gráfico
fig, ax = plt.subplots(figsize=(10, 6))
ax.set_facecolor(background_color)

bars = ax.bar(procedures_per_year.index, procedures_per_year.values, color=color_palette[:len(procedures_per_year)])
ax.set_title('Número de Procedimientos por Año (2018-2024)', fontsize=18, fontweight='bold', fontfamily='serif')
ax.set_xlabel('Año', fontsize=14, fontfamily='serif')
ax.set_ylabel('Número de Procedimientos', fontsize=14, fontfamily='serif')
ax.grid(color='#000000', linestyle=':', axis='y', zorder=0, dashes=(1, 5))

# Ocultar los bordes superiores, derechos e izquierdos
for s in ["top", "right", "left"]:
    ax.spines[s].set_visible(False)

# Añadir etiquetas a las barras
for bar in bars:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, yval + 0.05, int(yval), ha='center', va='bottom')

# Mostrar el gráfico
plt.show()
```

```python
median_duration_per_year = df_filtered.groupby('year')['duracion_proc'].median()
# Creación del gráfico
fig, ax = plt.subplots(figsize=(10, 6))
ax.set_facecolor(background_color)

bars = ax.barh(median_duration_per_year.index, median_duration_per_year.values, color=color_palette[:len(median_duration_per_year)])
ax.set_title('Mediana de Días de Duración de Procedimientos por Año (2018-2023)', fontsize=18, fontweight='bold', fontfamily='serif')
ax.set_ylabel('Año', fontsize=14, fontfamily='serif')
ax.set_xlabel('Mediana de Días de Duración', fontsize=14, fontfamily='serif')
ax.grid(color='#000000', linestyle=':', axis='x', zorder=0, dashes=(1, 5))

# Ocultar los bordes superiores, derechos e izquierdos
for s in ["top", "right", "left"]:
    ax.spines[s].set_visible(False)

# Añadir etiquetas a las barras
for bar in bars:
    xval = bar.get_width()
    ax.text(xval + 0.1, bar.get_y() + bar.get_height()/2, int(xval), ha='left', va='center')

# Mostrar el gráfico
plt.show()
```

```python
total_spent_per_year = df_filtered.groupby('year')['total_procedimiento'].sum()

# Creación del gráfico
fig, ax = plt.subplots(figsize=(10, 6))
ax.set_facecolor(background_color)

# Graficar líneas y puntos con colores específicos
for i, year in enumerate(total_spent_per_year.index):
    ax.plot(year, total_spent_per_year[year], marker='o', linestyle='-', color=color_palette[i], markersize=8)
    if i > 0:
        ax.plot(total_spent_per_year.index[i-1:i+1], total_spent_per_year.values[i-1:i+1], linestyle='-', color=color_palette[i], zorder=1)

ax.set_title('Gasto Total en Colones por Año (2018-2023)', fontsize=18, fontweight='bold', fontfamily='serif')
ax.set_xlabel('Año', fontsize=14, fontfamily='serif')
ax.set_ylabel('Gasto Total en Colones', fontsize=14, fontfamily='serif')
ax.grid(color='#000000', linestyle=':', axis='y', zorder=0, dashes=(1, 5))

# Ocultar los bordes superiores, derechos e izquierdos
for s in ["top", "right", "left"]:
    ax.spines[s].set_visible(False)

# Añadir la lista de totales en la parte superior derecha
textstr = '\n'.join([f'{year}: CRC {value:,.0f}' for year, value in total_spent_per_year.items()])
props = dict(boxstyle='round', facecolor='white', alpha=0.5)
plt.gcf().text(0.95, 0.95, textstr, fontsize=10, verticalalignment='top', bbox=props)

# Mostrar el gráfico
plt.show()
```

```python

procedure_type_counts = df['tipo_procedimiento'].value_counts()

procedure_type_percentages = procedure_type_counts / procedure_type_counts.sum()

procedure_type_counts = procedure_type_counts[procedure_type_percentages >= 0.005]
procedure_type_percentages = procedure_type_percentages[procedure_type_percentages >= 0.01]

max_index = procedure_type_percentages.idxmax()
max_value = procedure_type_percentages[max_index] * 100


fig, ax = plt.subplots(figsize=(12, 8), facecolor=background_color)
wedges, texts, autotexts = ax.pie(procedure_type_counts.values, colors=color_palette[:len(procedure_type_counts)], startangle=140, autopct='')


for i, wedge in enumerate(wedges):
    if procedure_type_counts.index[i] == max_index:
        angle = (wedge.theta2 + wedge.theta1) / 2
        x = wedge.r * 0.7 * np.cos(np.radians(angle))
        y = wedge.r * 0.7 * np.sin(np.radians(angle))
        ax.text(x, y, f'{max_value:.1f}%', horizontalalignment='center', verticalalignment='center', fontsize=14, fontfamily='serif', fontweight='bold', color='white')

ax.set_title('Frecuencia de Uso de Tipos de Procedimientos', fontsize=18, fontweight='bold', fontfamily='serif')

fig.patch.set_facecolor(background_color)

ax.legend(wedges, procedure_type_counts.index, title="Tipos de Procedimientos", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1), fontsize=12, title_fontsize=14, frameon=False)

plt.subplots_adjust(right=0.75)

plt.show()
```

## Antes de la pandemia

```python
total_gastado = df_periodo_0['total_procedimiento'].sum()

# Calcular la duración promedio de los procedimientos
duracion_promedio = df_periodo_0['duracion_proc'].mean()

# Calcular la media del total procedimiento
media_total_procedimiento = df_periodo_0['total_procedimiento'].median()

print(f"Total gastado en procedimientos: {total_gastado}")
print(f"Duración promedio de los procedimientos: {duracion_promedio}")
print(f"Promedio del total procedimiento: {media_total_procedimiento}")
```

```python
# Box Plot de duración de procedimientos
fig_box, ax_box = plt.subplots(figsize=(10, 6))
ax_box.set_facecolor(background_color)
sns.boxplot(data=df_periodo_0, x='duracion_proc', color=color_palette[0], ax=ax_box, showfliers=False)
ax_box.set_title('Duración de Procedimientos Antes de la Pandemia (2018-2019)', fontsize=14, fontweight='bold', fontfamily='serif')
ax_box.set_xlabel('Duración (días)', fontsize=14, fontfamily='serif')

for s in ["top", "right", "left"]:
    ax_box.spines[s].set_visible(False)

plt.show()
```

```python
# Bar Plot de tipo de procedimiento
fig_bar, ax_bar = plt.subplots(figsize=(10, 6))
ax_bar.set_facecolor(background_color)
tipo_procedimiento_counts = df_periodo_0['tipo_procedimiento'].value_counts().reset_index()
tipo_procedimiento_counts.columns = ['tipo_procedimiento', 'count']
sns.barplot(data=tipo_procedimiento_counts, x='tipo_procedimiento', y='count', palette=color_palette, ax=ax_bar)
ax_bar.set_title('Frecuencia de Tipos de Procedimiento Antes de la Pandemia (2018-2019)', fontsize=14, fontweight='bold', fontfamily='serif')
ax_bar.set_xlabel('Tipo de Procedimiento', fontsize=14, fontfamily='serif')
ax_bar.set_ylabel('Frecuencia', fontsize=14, fontfamily='serif')
ax_bar.grid(color='#000000', linestyle=':', axis='y', zorder=0, dashes=(1, 5))

for s in ["top", "right", "left"]:
    ax_bar.spines[s].set_visible(False)

plt.show()
```

```python
# Box Plot de duración de procedimientos
fig_box, ax_box = plt.subplots(figsize=(10, 6))
ax_box.set_facecolor(background_color)
sns.boxplot(data=df_periodo_2, x='duracion_proc', color=color_palette[5], ax=ax_box, showfliers=False)
ax_box.set_title('Duración de Procedimientos Después de la Pandemia (2022-2024)', fontsize=14, fontweight='bold', fontfamily='serif')
ax_box.set_xlabel('Duración (días)', fontsize=14, fontfamily='serif')

for s in ["top", "right", "left"]:
    ax_box.spines[s].set_visible(False)

plt.show()
```

```python
# Bar Plot de tipo de procedimiento
fig_bar, ax_bar = plt.subplots(figsize=(10, 6))
ax_bar.set_facecolor(background_color)

# Obtener las 5 categorías más frecuentes
tipo_procedimiento_counts = df_periodo_2['tipo_procedimiento'].value_counts().nlargest(5).reset_index()
tipo_procedimiento_counts.columns = ['tipo_procedimiento', 'count']

# Crear el gráfico de barras
sns.barplot(data=tipo_procedimiento_counts, x='tipo_procedimiento', y='count', palette=color_palette, ax=ax_bar)
ax_bar.set_title('Frecuencia de Tipos de Procedimiento Después de la Pandemia (2022-2024)', fontsize=14, fontweight='bold', fontfamily='serif')
ax_bar.set_ylabel('Frecuencia', fontsize=14, fontfamily='serif')
ax_bar.grid(color='#000000', linestyle=':', axis='y', zorder=0, dashes=(1, 5))

# Rotar las etiquetas del eje x
ax_bar.set_xticklabels(ax_bar.get_xticklabels(), rotation=45, horizontalalignment='right', fontsize=12, fontfamily='serif')

for s in ["top", "right", "left"]:
    ax_bar.spines[s].set_visible(False)

plt.show()
```
